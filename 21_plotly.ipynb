```python
import pandas as pd
import numpy as np
import datetime
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# CSV形式で保存したセンサーデータをデータフレームに読み込む（index_col=0で1列目をインデックスに指定）
df = pd.read_csv(r"E:\work\suds_sig6\omron_samples\csv_files\CA43F0B62495_interval_60.csv", index_col=0)

# データフレームの先頭5行を表示
print(df.head())

# グラフに表示するデータの期間を指定
# df = df["2024-05-10 00:00:00":"2024-05-12 23:59:59"]
# 期間指定後のデータの先頭5行を表示
# print(df.head())
```

```python
# グラフ表示用のインスタンスを作成
# graph_data = ['temparature', 'relative_humidity']
graph_data = ['temparature', 'relative_humidity', 'ambient_light', 'barometric_pressure', 'sound_noise', 'eTVOC', 'eCO2']
fig = make_subplots(rows=len(graph_data), cols=1, subplot_titles=graph_data, 
                    shared_xaxes='all', 
                    vertical_spacing=0.01)

# グラフを表示
for i in range(len(graph_data)):
    fig.add_trace(go.Scatter(x=df.index, y=df[graph_data[i]], mode='lines', name=graph_data[i]), row=i+1, col=1)

# グラフが1つの時は、レンジスライダーを表示すると表示範囲の指定が容易
# fig.update_xaxes(rangeslider_visible=True)

fig.update_layout(title='Omron Sensor Data', 
                #   xaxis_title='Time', 
                  yaxis_title='Sensor Data',
                  showlegend=True,
                  width=1000,
                  height=400*len(graph_data),
                )
fig.show()
```
